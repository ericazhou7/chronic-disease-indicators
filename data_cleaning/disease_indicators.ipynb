```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('U.S._Chronic_Disease_Indicators__CDI_.csv')
```

```python
df = df[df['Topic']=='Overarching Conditions']
df = df[['YearStart', 'LocationAbbr','Question','DataValueUnit','DataValueType','DataValue','DatavalueFootnote',
'LowConfidenceLimit','HighConfidenceLimit','Stratification1','StratificationCategory1','GeoLocation']]
df = df.reset_index(drop=True)
```

```python
df_us = df[df['LocationAbbr']=='US'][['YearStart','Question','DataValueType','DataValue',
                                      'Stratification1']].reset_index(drop=True)
```

```python
for row in df_us.itertuples():
    df.loc[(df.YearStart == row.YearStart) & (df.Question == row.Question) & (df.DataValueType == row.DataValueType)
             & (df.Stratification1 == row.Stratification1),'US_Avg'] = row.DataValue
```

```python
df['Diff_from_US'] = df.DataValue.astype(float) - df.US_Avg.astype(float)
```

```python
df[['Latitude','Longitude']] = df['GeoLocation'].str[1:-1].str.split(',', expand=True).astype(float)
df.drop('GeoLocation',axis=1)
```
